```python
titanic = pd.read_csv('titanic.csv')
titanic
```

Calculate the average age of passeners
(1 point)

```python
titanic.Age.mean()
```

Calculate the percentage of male and female passengers
(1 point)

```python
n_all = len(titanic.index)  # number of all passengers
n_male =  len(titanic[titanic.Sex == 'male'])
n_female =  len(titanic[titanic.Sex == 'female'])
print(f'male: {n_male/n_all*100:.1f}%, female: {n_female/n_all*100:.1f}%')
```

Create a DataFrame of passenger who survived
(1 point)

```python
titanic_survived =  titanic.loc[titanic.Survived == True]
titanic_survived
```

What's the average age of survived passengers compared to those who did not survive?
(1 point)

```python
mean_age_survived =  titanic_survived.Age.mean()
titanic_not_survived = titanic.loc[titanic.Survived == False]
mean_age_did_not_survive = titanic_not_survived.Age.mean()
print(f'mean_age_survived: {mean_age_survived:.1f}, mean_age_did_not_survive: {mean_age_did_not_survive:.1f}')
```

Rename passengers who did not survive to 'anonymous'
(2 points)

```python
titanic_renamed = titanic.copy()
titanic_renamed.loc[titanic_renamed.Survived == False, 'Name'] = 'anonymous'
```

```python
titanic_renamed
```

Calculate average fare for passenger class (Pclass) 1, 2, and 3
(2 points)

```python
class_1 = titanic.loc[titanic.Pclass == 1]
class_2 = titanic.loc[titanic.Pclass == 2]
class_3 = titanic.loc[titanic.Pclass == 3]
```

```python
avg_class_1 = class_1.Fare.mean()
avg_class_2 = class_2.Fare.mean()
avg_class_3 = class_3.Fare.mean()
print(
    f'Average fare by class - 1st Class: {avg_class_1:.1f}, 2nd Class: {avg_class_2:.1f}, 3rd Class: {avg_class_3:.1f}')
```

Calculate the percentage of passengers who survived in each class
(2 points)

```python
n_1 = len(class_1)
n_2 = len(class_2)
n_3 = len(class_3)

n_survived_1 = class_1.Survived.sum()
n_survived_2 = class_2.Survived.sum()
n_survived_3 = class_3.Survived.sum()

print(f'Survival by class - 1st Class: {n_survived_1/n_1*100:.1f}%, 2nd Class: {n_survived_2/n_2*100:.1f}%, 3rd Class: {n_survived_3/n_3*100:.1f}%')
```
