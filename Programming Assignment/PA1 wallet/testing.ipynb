```python
def gsh(target, arr, index=0):

    if target == 0:
        return []

    if target < 0 or index == len(arr):
        return None

    prev_rec = gsh(target-arr[index], arr, index+1)
    #* TARGET = 5, [1, 2, 3]
    #*              ^
    #* TARGET = 5-1, [1, 2, 3]
    #*                   ^

    if prev_rec is None:
        return gsh(target, arr, index+1)

    return [arr[index]] + prev_rec
```

```python

def withdraw(self, amount):
    if amount.value < 0:
            raise ValueError('A value should be positive.')
        if not isinstance(amount, object):
            raise TypeError("An amount should be a money object.")
        # A recursion step but with a copied items list
        towithdraw = withdraw_from(amount, self.get_items)
        if towithdraw != None:
            for val in towithdraw:
                self.items.remove(val)
        return towithdraw


def withdraw_from(amount, items):
    if amount == 0:
            return []
    if amount != 0 and items == []:
        return None
    
    first = items[0]
    remain = items[1:]
    if amount.currency == first.currency:
        if amount.value >= first.value: # check that positive
            rest = withdraw_from(amount-first,items)
            if rest is not None:
                return [first] + rest
    return withdraw_from(amount,remain)
```
