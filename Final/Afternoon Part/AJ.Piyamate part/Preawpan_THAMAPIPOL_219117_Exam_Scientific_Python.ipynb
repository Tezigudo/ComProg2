```python
dinner = restaurant[restaurant.time == 'Dinner']
launch = restaurant[restaurant.time == 'Lunch']
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

dinner.hist(ax=axes[0], column='tip')
launch.hist(ax=axes[1], column='tip')
axes[0].set_title('Dinner')
axes[1].set_title('Lunch')

time_list = ['Dinner', 'Lunch']

for i in range(2):
    axes[i].set_xlabel(f'tip amount for {time_list[i]}')
    axes[i].set_ylabel('number of bills')

fig.suptitle('Histogram of tip amount')
```

# COVID-19 cases

## 1) COVID cases data import from JHU (1 point)

Import worldwide **daily** COVID dataset provided by Johns Hopkins University
from https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/jhu/full_data.csv

```python
covid = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/jhu/full_data.csv', date_parser='date')

covid
```

## 2) COVID cases in Thailand (2 points)

Plot the number of total cases for Thailand over time (x-axis: date, y-axis: number of cases **in million**) using **area plot**, set labels for x & y axes and plot title appropriately

Hint: `pivot()`, `dataFrame.column.plot.area()`

```python
covid.pivot('date', 'location', 'total_cases')['Thailand'].plot.area(ylabel='total cases', title='Total cases of covid-19 in Thailand')
```

## 3) COVID cases globally (1 point)

Plot the number total cases globally over time (x-axis: date, y-axis: number of cases **in billion**) using **area plot**, set labels for x & y axes and plot title appropriately

```python
covid.pivot('date', 'location', 'total_cases')['World'].plot.area(ylabel='total cases', title='Total global cases of covid-19')
```

## 4) COVID cases comparison #1 (4 points)

Plot number of cases over time for the following countries (x-axis: date, y-axis: number of cases **in million**) using **line plot**
*   Thailand
*   United Kingdom
*   China
*   United States
*   Singapore
*   New Zealand

Conditions:
*   Put all 4 data lines in one plot
*   Show data legend
*   Use log scale for y-axis
*   Set figure size to 20 x 10 inch
*   Set labels for x & y axes and plot title appropriately
*   Show gridlines

Hints:
*   `plt.legend()`
*   `plot(figsize=(20,10, logy=True)`
*   `plot.grid(True)`


```python
covid_copy = covid.copy()
covid_copy['total_cases'] /= 1e6
data = covid_copy.pivot('date', 'location', 'total_cases')[['Thailand', 'United Kingdom', 'China', 'United States']]
data.plot(figsize=(20, 10), logy=True, grid=True, ylabel='number of cases in million', title='number of COVID cases over time at Thailand, United Kingdom, China and United States')
```

## 5) COVID cases comparison #2 (3 points)

Plot the same data above, but convert the data in y axis from `number of cases` to `per million population`. Use population data provided by this file: https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/jhu/locations.csv

```python
population = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/jhu/locations.csv')
population
```

```python
merged_data = covid.merge(population, on='location')
# merged_data.set_index('date', inplace=True)
```

```python
merged_data['cases_per_million_population'] = merged_data['total_cases'] / (merged_data['population'] / 1e6)
merged_data
```

```python

data = merged_data.pivot('date', 'location', 'cases_per_million_population')[['Thailand', 'United Kingdom', 'China', 'United States']]
data.plot(figsize=(20, 10), logy=True, grid=True, ylabel='number of cases per million population', title='number of COVID cases per million population over time at Thailand, United Kingdom, China and United States')
```

## 6) COVID new cases per day comparison (3 points)

Using the same format as above, plot `number of new cases per day` in `per million population`. Because the daily data is somewhat noisy, use a moving average of 14-day period to smoothen the data.

Hint: `covid_new_cases_percent['Thailand'].rolling(window=14).mean().plot()`

```python
merged_data['new_cases_per_day_million_population'] = merged_data['new_cases'] / (merged_data['population'] / 1e6)

merged_data
```

```python
data = merged_data.pivot('date', 'location', 'new_cases_per_day_million_population')[['Thailand', 'United Kingdom', 'China', 'United States']].rolling(window=14).mean()
data.plot(figsize=(20, 10), logy=True, grid=True, ylabel='14 day average of new cases per day permillion population', title='14 days average of number of new COVID cases per day over time at Thailand, United Kingdom, China and United States')
```
